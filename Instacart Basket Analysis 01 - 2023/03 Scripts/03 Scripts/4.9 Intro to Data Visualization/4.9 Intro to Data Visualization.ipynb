## Creating bar charts

```python
# Creating bar charts
df_ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

## Exporting charts

```python
# Saving chart
bar = df_ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar(color = ['darkblue', 'darkslategray', 'navy','steelblue', 'deepskyblue', 'turquoise', 'royalblue'])
```

```python
# Calling chart
bar.figure
```

```python
# Exporting bar chart 
bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_orders_dow.png'))
```
