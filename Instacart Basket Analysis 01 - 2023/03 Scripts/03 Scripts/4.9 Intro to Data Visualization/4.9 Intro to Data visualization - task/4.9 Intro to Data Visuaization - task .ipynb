```python
# Cheking mixed data types
for col in df_customers_clean.columns.tolist():
  weird = (df_customers_clean[[col]].applymap(type) != df_customers_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers_clean[weird]) > 0:
    print (col)
```
